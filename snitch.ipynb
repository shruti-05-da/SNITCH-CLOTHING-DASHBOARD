### clean dataset

```python
snitch
```

```python
snitch.columns=snitch.columns.str.lower()
```

```python
snitch["units_sold"].fillna(round(snitch["profit"]/snitch["unit_price"]),inplace=True)
```

```python
snitch
```

```python
snitch.dropna(subset=["order_date"], inplace=True)
```

```python
snitch.info()
```

```python
snitch["sales_amount"]=snitch["unit_price"]-(snitch["unit_price"]*snitch["discount_%"])
```

```python
snitch.info()
```

```python
snitch.dropna(subset=["unit_price", "discount_%", "segment"],how="all", inplace=True)
```

```python
snitch
```

```python
snitch.info()
```

```python
snitch["discount_%"].fillna(0,inplace=True)
```

```python
snitch["sales_amount"].fillna(snitch["unit_price"],inplace=True)
```

```python
snitch.info()
```

```python
snitch["segment"].fillna("none",inplace=True)
```

```python
snitch.dropna(subset="units_sold",inplace=True)
```

```python
snitch.info()
```

```python
snitch["unit_price"].fillna(0,inplace=True)
```

```python
snitch["sales_amount"].fillna(0,inplace=True)
```

```python
snitch
```

```python
snitch.info()
```

```python
snitch.drop(columns="order_id",inplace=True)
```

```python
snitch
```

```python
snitch.reset_index(drop=True, inplace=True)
```

```python
snitch
```

```python
snitch["units_sold"]=snitch["units_sold"].astype(int)
```

```python
snitch
```

```python
snitch["loss"] = snitch["profit"].where(snitch["profit"] < 0, other=0)
```

```python
snitch["loss"] = snitch["loss"].abs()
```
